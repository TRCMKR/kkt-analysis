```python
import numpy as np
from sklearn.linear_model import LinearRegression
from matplotlib import pyplot as plt
```

```python
def extrapolate(x, y, x_from, x_to):
    lm = LinearRegression()
    lm.fit(np.array(x).reshape(-1, 1), y)

    test_x = np.linspace(x_from, x_to, 100).reshape(-1, 1)
    test_y = lm.predict(test_x)

    lm2 = LinearRegression()
    lm2.fit(np.log(np.array(x)).reshape(-1, 1), y)

    test_y2 = lm2.predict(np.log(test_x))

    plt.figure()
    plt.plot(x, y, 'o', label='Given Data')
    plt.plot(test_x, test_y, label='Linear Model')
    plt.plot(test_x, test_y2, label='Log-Linear Model')
    plt.legend()
    plt.show()

    return lm, lm2
```

```python
x = [7, 8, 9]
y = [11, 16, 20]

lm, lm2 = extrapolate(x, y, 7, 20)
```

```python
y_pred = lm2.predict(np.log(np.array([[17]])))
print(y_pred)
```
