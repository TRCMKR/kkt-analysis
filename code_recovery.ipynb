```python
r = 7
m = 17
d = 2 ** (m - r)
t = d // 3
delta = d - 2 * t - 1
# delta = 60
# t = (d - 1 - delta) // 2

n = 2 ** m
k = sum([comb(m, i) for i in range(0, r + 1)])
print(n, k, t, delta)
```

```python
def ro(p, l):
    top = comb((k + l) // 2, p // 2) * comb((k + l) // 2, p // 2) * comb(n - k - l, t - p)
    return top / comb(n, t)
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from matplotlib import pyplot as plt
```

```python
x = [7, 8, 9]
y = [11, 16, 20]

lm = LinearRegression()
lm.fit(np.array(x).reshape(-1, 1), y)

test_x = np.linspace(7, 20, 100).reshape(-1, 1)
test_y = lm.predict(test_x)

lm2 = LinearRegression()
lm2.fit(np.log(np.array(x)).reshape(-1, 1), y)

test_y2 = lm2.predict(np.log(test_x))

plt.figure()
plt.plot(x, y, 'o', label='Given Data')
plt.plot(test_x, test_y, label='Linear Model')
plt.plot(test_x, test_y2, label='Log-Linear Model')
plt.legend()
plt.show()
```

```python
y_pred = lm2.predict(np.log(np.array([[17]])))
print(y_pred)
```
